```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

import moon_classification_problem as mcp
from moon_classification_problem import moon_plot
```

```python
plt.ioff()
```

```python
# An ordinary least square (OLS) filter algorithm
zero_separation_moon = mcp.MoonProblem(d=0)
training_sample = zero_separation_moon.get_training_sample()
test_sample = zero_separation_moon.get_test_sample()
(xa, ya), (xb, yb) = training_sample
X = np.hstack((np.vstack((np.ones(xa.size), xa, ya)), np.vstack((np.ones(xb.size), xb, yb)))).T
Y = np.hstack((np.ones(xa.size), -np.ones(xb.size)))
Rxx = X.T @ X
rdx = X.T @ Y
Wml = np.linalg.inv(Rxx) @ rdx
w = Wml
```

```python
fig = plt.figure()
ax = moon_plot(fig.add_subplot(111), test_sample)
x = np.linspace(-20, 40, 50)
ax.plot(x, -(w[0] + w[1] * x) / w[2], 'black')
plt.show()
```
